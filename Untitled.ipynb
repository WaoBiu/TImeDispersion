```python
from Modeling import *
from TDE import Source_Correctiion, Wave_Correctiion
```

```python
tmax = 1.5
nt = 1501
dt = 0.001

xmax = 1000
nx = 201
dx = 5

zmax = 1000
nz = 201
dz = 5

pmodel = np.ones((nx, nz), dtype=np.float32) * 2500
vpmodel = np.ones((nx, nz), dtype=np.float32) * 2000
vsmodel = np.ones((nx, nz), dtype=np.float32) * 1600

pmodel[:,150:] = 3000
vpmodel[:,150:] = 2500
vsmodel[:,150:] = 2000

pml_len = 10
```

```python
test_model = FirstOlder_NPML_PS(pmodel, vpmodel, vsmodel, tmax, xmax, zmax, nt, nx, nz, pml_len)
```

```python
t_0 = 0.05
f_m = 30
source_position = [11, 100]  # (z , x)
source_time_array = np.linspace(0, 0.1, 101)
original_source = 1 * (-4 * (pi * f_m * (source_time_array-t_0)) * pi * f_m * \
                         np.exp(-1 * ( pi * f_m * (source_time_array-t_0))**2) + \
                         (1 - 2 * (pi * f_m * (source_time_array-t_0))**2) * \
                         np.exp(-1 * ( pi * f_m * (source_time_array-t_0))**2) * \
                            -2 * (pi * f_m * (source_time_array-t_0)) * pi * f_m)
```

```python
plt.plot(original_source)
plt.title("source")
plt.show()
```

```python
modify_source = Source_Correctiion(source_time_array, dt, original_source, source_time_array, dt)
max_source = max(modify_source.max(), abs(modify_source.min()))
modify_source = modify_source / max_source
```

```python
plt.plot(modify_source)
plt.title("source")
plt.show()
```

```python
test_model.o2xFM(wavelet=modify_source, wavalet_position=source_position, wavalet_direction='z')
```
