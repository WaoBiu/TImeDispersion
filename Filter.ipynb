```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, pi, sin
from scipy.special import jv
```

```python
def sinc(x):
    if x == 0:
        return 1
    else:
        return sin(x)/x
```

```python
def Set_TD_Eli_Filter(n, m):
    td_eli_filter = np.zeros((n,m), dtype=float)
    for i in range(n):
        for j in range(m):
            # sinc_item = sinc(pi*(j-k/2))
            for k in range(-10000,10001):
                td_eli_filter[i,j] += jv(k+1,2*i) * sinc(pi*(j-k/2)) / (1-(j-k/2)+pow(10,-7))
    return td_eli_filter * 0.5

def Set_TD_Add_Filter(n,m):
    td_add_filter = np.zeros((n,m), dtype=float)
    for i in range(n):
        for j in range(m):
            # sinc_item = sinc(pi*(j-k/2))
            for k in range(-10000,10001):
                td_add_filter[i,j] += sinc(pi*(i-k/2)) * jv(k,2*j)
    return td_add_filter  
```

```python
n = m = 1000
a = Set_TD_Eli_Filter(n,m)
b = Set_TD_Add_Filter(n,m)
```

```python
plt.figure(figsize=(10,10))
plt.title("Eliminate Filter")
plt.imshow(a,interpolation='none')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title("Add Filter")
plt.imshow(b,interpolation='none')
plt.colorbar()
plt.show()
```

```python
s = np.dot(a,b)
```

```python
plt.figure(figsize=(10,10))
plt.title("Dot Filter")
plt.imshow(s,interpolation='none')
plt.colorbar()
plt.show()
```

```python
zp = np.zeros((n,m))
for i in range(n):
    for j in range(m):
        zp[i,j] = 2 * sinc(2*(i-j)) / pi
```

```python
plt.figure(figsize=(10,10))
plt.title("Should Filter")
plt.imshow(zp,interpolation='none')
plt.colorbar()
plt.show()
```

```python
diff = zp - s
```

```python
plt.figure(figsize=(10,10))
plt.title("Diff Filter")
plt.imshow(diff,interpolation='none')
plt.colorbar()
plt.show()
```
