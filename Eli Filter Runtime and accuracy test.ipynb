```python
def sinc(x):
    if x == 0:
        return 1
    else:
        return sin(x)/x
```

# When n=m=10

```python
def Set_TD_Eli_Filter(n, m):
    td_eli_filter = np.zeros((n,m), dtype=float)
    for i in range(n):
        for j in range(m):
            # sinc_item = sinc(pi*(j-k/2))
            for k in range(-10,11):
                td_eli_filter[i,j] += jv(k+1,2*i) * sinc(pi*(j-k/2)) / (1-(j-k/2)+pow(10,-7))
    return td_eli_filter * 0.5
```

```python
start = time.time()
n = m = 10
a = Set_TD_Eli_Filter(n,m)
end = time.time()
plt.figure(figsize=(5,5))
plt.title("k = 10")
plt.imshow(a,interpolation='none')
plt.colorbar()
plt.show()
print(end-start)
```
